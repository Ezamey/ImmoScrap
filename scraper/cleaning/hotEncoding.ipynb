# ^ Type de données propices à un "Hot encoding"

##  data["heating_type_test"]

```python
data["heating_type_test"] = pd.Categorical(data["heatingType"])
```

```python
data_dummies = pd.get_dummies(data["heating_type_test"],prefix="heating_with")
```

```python
del data["heatingType"]
```

```python
data =pd.concat([data,data_dummies],axis=1)
```

## data["condition"]

```python
data["condition_test"]=pd.Categorical(data["condition"])
```

```python
data_dummies = pd.get_dummies(data["condition_test"],prefix="condition_is ")
```

```python
del data["condition"]
```

```python
data = pd.concat([data,data_dummies],axis=1)
```

```python
data["type_test"] = pd.Categorical(data["type"])
```

```python
data_dummies = pd.get_dummies(data["type_test"],prefix="is")
```

```python
del data["type"]
```

```python
data = pd.concat([data,data_dummies],axis=1)
```

## data["subtype"]

```python
data["subtype_test"] = pd.Categorical(data["subtype"])
```

```python
data_dummies = pd.get_dummies(data["subtype_test"],prefix="subtype_is")
```

```python
del data["subtype"]
```

```python
data = pd.concat([data,data_dummies],axis=1)
```

```python
del data["condition_test"]
del data["type_test"]
del data["heating_type_test"]
del data["subtype_test"]
```
